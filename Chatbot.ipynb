```python
import keras
import os
import gensim
import nltk
import yaml
import pickle
import numpy as np
from gensim import corpora,models,similarities
from gensim.models import Word2Vec
from sklearn.model_selection import train_test_split
from keras.models import Sequential
from keras.layers.recurrent import LSTM
```

```python

file = open('ai.yml','rb')
data = yaml.safe_load(file)
con = data['conversations']
```

```python
q = []
a = []
for i in range(len(con)):
    for j in range(len(con[i])):
        if j<len(con[i])-1:
            q.append(con[i][j])
            a.append(con[i][j+1])
```

```python
nltk.download()
```

```python
tok_q = []
tok_a = []
for i in range(len(q)):
    tok_q.append(nltk.word_tokenize(q[i].lower()))
    tok_a.append(nltk.word_tokenize(a[i].lower()))

pad = np.ones((32,),dtype = np.float32)
```

```python
vocab = tok_q+tok_a
model = gensim.models.Word2Vec(vocab,min_count = 1, size=32)
#print(tok_q[1])
```

```python
vec_q = []
for i in tok_q:
    sentvec = [model[w] for w in i]
    vec_q.append(sentvec)
#print(vec_q[1])
```

```python
vec_a = []
for i in tok_a:
    sentvec = [model[w] for w in i]
    vec_a.append(sentvec)
```

```python
for d in vec_q:
    d[14:] = []
    d.append(pad)
#print(vec_q[1])
for d in vec_q:
    if len(d)<15:
        for i in range (15-len(d)):
            d.append(pad)
#print(vec_q[1])
```

```python
for d in vec_a:
    d[14:] = []
    d.append(pad)
#print(vec_a[1])
for d in vec_a:
    if len(d)<15:
        for i in range (15-len(d)):
            d.append(pad)
#print(vec_a[1])
```

```python
vec_q = np.array(vec_q,dtype = np.float64)
vec_a = np.array(vec_a,dtype = np.float64)
q_train,q_test,a_train,a_test = train_test_split(vec_q,vec_a,test_size=0.2,random_state=1)
```

```python
print(q_train.shape[1:])
```

```python
model = Sequential()
model.add(LSTM(output_dim=32,input_shape=q_train.shape[1:],return_sequences=True,init='glorot_normal',inner_init='glorot_normal',activation='sigmoid'))
model.add(LSTM(output_dim=32,input_shape=q_train.shape[1:],return_sequences=True,init='glorot_normal',inner_init='glorot_normal',activation='sigmoid'))
model.add(LSTM(output_dim=32,input_shape=q_train.shape[1:],return_sequences=True,init='glorot_normal',inner_init='glorot_normal',activation='sigmoid'))
model.add(LSTM(output_dim=32,input_shape=q_train.shape[1:],return_sequences=True,init='glorot_normal',inner_init='glorot_normal',activation='sigmoid'))
model.compile(loss='cosine_proximity',optimizer = 'adam', metrics=['accuracy'])
```

```python
model.fit(q_train,a_train,epochs=500,validation_data=(q_test,a_test))
```
